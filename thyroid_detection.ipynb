```python
name0=df['target'].unique().tolist()
name=sorted(name0)
print(name)
N=list(range(len(name)))
normal_mapping=dict(zip(name,N)) 
reverse_mapping=dict(zip(N,name))
df['target']=df['target'].map(normal_mapping)
```
